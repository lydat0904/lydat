```python
#W4A9
def scp (n) :
    if n < 0 :
        return False
    k = int ( n ** 0.5 )
    if k * k == n :
        return True 
    return False 
def chuso (n):
    s= str (n)
    for i in  range (len(s)):
        for j in range ( i + 1, len(s)): 
            if s[i] == s[j] :
                return False
    return True
n = int ( input ())
for i in range (1 , n+1) :
    if scp (i) == True and chuso(i) == True :
        print (i, end = " ")
```

```python
#W4A10
def collatz(n) :
    step = 1
    while n != 1 :
        if n % 2 == 0:
            n = n// 2
        else:
            n = 3 * n + 1
        step += 1
    return step
def timso (n) :
    max = -1
    somin = 1e9
    for i in range ( 1, n + 1) :
        step = collatz(i)
        if step > max or ( step == max and i < somin) :
            max = step
            somin = i
    return somin , max
n =int ( input ())
print (timso (n))
```

```python
#W4A13
def tonguoc (n):
    sum = 0
    for i in range (1, int (n**0.5)+1):
        if n % i == 0:
            sum += i
            if i != n // i:
                sum += n // i
    return sum - n
a , b = map (int, input().split())
if tonguoc (a) == b and tonguoc (b) == a:
    print ("YES")
else:
    print ("NO")
```
