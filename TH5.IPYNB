```python
#W5A3
def snt (n) :
    if n < 2 :
        return False 
    else :
        for  i in range ( 2 , int (n ** 0.5) +1 ):
            if n % i == 0 :
                return False 
    return True
n  = int ( input())
if snt (n) :
    print ( n , "la so nguyen to ")
else :
    print ( n , "khong phai so nguyen to ")
```

```python
#W5A7
def casio(a , b , op) :
    if op == "+" :
        return a + b 
    elif op == "-" :
        return a - b
    elif op == "*" :
        return a * b
    elif op == "/" :
        c = f"{(a/b) :.2f}"
        return c
a , b , op = map (str,input().split())
a = int ( a )
b = int ( b )
print ( casio ( a , b , op ))


    
```
