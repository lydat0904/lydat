```python
#W7A27
def ngoac_dung(n):
    a = []
    b = {")":"(","]":"[","}":"{"}
    for char in n:
        if char in b.values():
            a.append(char)
        elif char in b.keys():
            if len(a) == 0 or a.pop() != b[char]:
                return False
    return len(a) == 0
n = input()
if ngoac_dung(n):
    print("YES")
else:
    print("NO")
```

```python
#W7A29
def tim_day(n):
    a = len(n)
    if a == 0:
        return []
    dp = [1] * a #độ dài dãy tăng dài nhất tại vị trí i, bau đầu coi mỗi phần tử là 1 dãy tăng với độ dài là 1
    vet = [-1] * a #phần tử đứng ngay trước n[i] trong dãy con, -1 là không có phần tử nào đứng trước

    for i in range(a):
        for j in range(i):
            if n[i] > n[j] and dp[i] < dp[j] + 1:
                dp[i] = dp[j] + 1
                vet[i] = j
    max_len = 0
    end_index = -1

    for i in range(a):
        if dp[i] > max_len:
            max_len = dp[i]
            end_index = i
    
    result = []
    curr = end_index
    while curr != -1:
        result.append(n[curr])
        curr = vet[curr]
    return result[::-1]

a = list(map(int,input().split()))
print(tim_day(a))


```

```python
#W7A29
from bisect import bisect_left
def tim_day(nums):
    n = len(nums)
    if n == 0:
        return []
    
    parent = [-1] * n
    tails = []
    tails_indices = []

    for i,x in enumerate(nums):
        idx = bisect_left(tails,x)
        if idx == len(tails):
            tails.append(x)
            tails_indices.append(i)
        else:
            tails[idx] = x
            tails_indices[idx] = i
        if idx > 0:
            parent[i] = tails_indices[idx - 1]
            
    curr_index = tails_indices[-1]
    result = []
    while curr_index != -1 :
        result.append(nums[curr_index])
        curr_index = parent[curr_index]
    
    return result[::-1]
n = list(map(int,input().split()))
print(tim_day(n))
```
