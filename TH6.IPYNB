```python
#w6a9
def input_dict (a) :
    dict_a = {}
    k = a.split()
    for i in k :
        try:
            key , value = i.split(':')
        except ValueError :
            continue
        key =key.strip()
        value=value.strip()
        if key in dict_a :
            dict_a[key].append(value)
        else :
            dict_a[key] = [value]
    return dict_a
def xu_ly (a , b) :
    c = {}
    for key in a :
        c[key]=a[key]
    for key in b :
        if key in c :
            c[key]+=b[key]
            c[key].sort()
        else :
            c[key]=b[key]

    return c
a = input()
b = input()


print (xu_ly(input_dict(a), input_dict(b)))
```

```python
#w6a13
def tachkey_value (n) :
    dict_n = {}
    k = n.split()
    for i in k :
        try :
            key , value = i.split(":")
        except ValueError :
            continue
        key =key.strip()
        value = value.strip()
        dict_n[key]=value
    return dict_n
def dao_dict(n) :
    newdict_n ={}
    for key in n :
        newdict_n[n[key]] = key
    return newdict_n
n = input()
print(dao_dict(tachkey_value(n)))
```

```python
#w6a15
def input_dict (n) :
    dict_n = {}
    k = n.split()
    for i in k :
        try :
            key ,value = i.split(':')
            key = key.strip()
            value = value.strip()
            value_int = int(value)
        except ValueError :
            continue
        if key not in dict_n :
            dict_n[key] = value_int
    return dict_n
def xu_ly(n , k) :
    dict_ans ={}
    for i in n :
        if n[i] > k :
            dict_ans[i] = n[i]
    return dict_ans
n = input()
k = int (input())
print(xu_ly(input_dict(n),k))
```
